```python
# Predicting USD/PLN exchange rate with LSTM neural network
# Imports
import os
import math
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import mean_squared_error as mse
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
# Disabling debugging logs for tensorflow - info about deprecated functions will not be printed
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
import tensorflow as tf
tf.logging.set_verbosity(tf.logging.ERROR)
```

```python
# Setting CUDA_VISIBLE_DEVICES to -1 - it will ensure that tensorflow is run with CPU
os.environ['CUDA_VISIBLE_DEVICES'] = '-1'
```

```python
# Reading dataset
dataset = pd.read_csv('usd_pln_historic.csv')
# Creating np.array from the dataset
rates_values = dataset.iloc[:, 3:4].values
```

```python
# Feature scaling
scaler = MinMaxScaler(feature_range=(0, 1))
rates_values_scaled = scaler.fit_transform(rates_values)
```

```python
# setting how many previous days are taken into single prediction
interval = 100
# setting the amount of days for which exhange rate will be predicted
days_to_predict = 300
```

```python
# getting the usd/pln exchange rates values for both training and test sets
training_values = rates_values_scaled[0:-days_to_predict, :]
test_values = rates_values_scaled[-days_to_predict:, :]
```

```python
# creating test set - it must have interval + days_to_predict values, since each new prediction require previous interval values
inputs = rates_values_scaled[-days_to_predict-interval:, :]
X_test = []
```

```python
# appending values to test set
for i in range(interval, days_to_predict + interval):
    X_test.append(inputs[i-interval:i, 0])
```

```python
# reshaping X_test so it will be compatible Keras - from docs: 3D tensor with shape (batch_size, timesteps, input_dim)
X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
```

```python
# creating training set and appending data
X_train = []
y_train = []
for i in range(interval, len(training_values)):
    X_train.append(training_values[i-interval:i, 0])
    y_train.append(training_values[i, 0])
```

```python
# converting lists to np arrays
X_train, y_train = np.array(X_train), np.array(y_train)
```

```python
# reshaping corresponding to have a format required by Keras - from docs: 3D tensor with shape (batch_size, timesteps, input_dim).
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
```

```python
# Initializing RNN
regressor = Sequential()
```

```python
# First layer with dropout regularisation
regressor.add(LSTM(units = 50, return_sequences=True))
regressor.add(Dropout(0.2))
```

```python
# Second layer with dropout regularisation
regressor.add(LSTM(units = 50, return_sequences=True))
regressor.add(Dropout(0.2))
```

```python
# Third layer with dropout regularisation
regressor.add(LSTM(units = 50, return_sequences=True))
regressor.add(Dropout(0.2))
```

```python
# Fourth layer with dropout regularisation
regressor.add(LSTM(units = 50, return_sequences=True))
regressor.add(Dropout(0.2))
```

```python
# Fifth layer with dropout regularisation
regressor.add(LSTM(units = 50, return_sequences=True))
regressor.add(Dropout(0.2))
```

```python
# Sixth layer with dropout regularisation
regressor.add(LSTM(units = 50, return_sequences=False))
regressor.add(Dropout(0.2))
```

```python
# Output layer
regressor.add(Dense(units = 1))
```

```python
# Compiling RNN 
regressor.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Training RNN
history = regressor.fit(X_train, y_train, epochs = 150, batch_size=32)
```

```python
# Getting predicted exchange values
predicted_exchange = regressor.predict(X_test)
predicted_exchange = scaler.inverse_transform(predicted_exchange)
test_values = scaler.inverse_transform(test_values)
```

```python
# Getting Root Mean Squared Error (RMSE)
rmse = math.sqrt(mse(y_true = test_values, y_pred = predicted_exchange))
```

```python
# Get Standard Deviation for the whole database
sd = np.std(rates_values)
```

```python
# Printing statistics
diff = math.sqrt(math.pow(sd - rmse, 2))
print(f"RMSE for predictions is equal to:  {rmse:.2f} PLN.\n"
      f"Standard deviation for the whole dataset is equal to: {sd:.2f} PLN\n"
      f"Difference between standard deviation and predicted result is equal to: {diff:.2f} PLN")
```

```python
# Visualising loss function
training_loss = history.history['loss']
epoch_count = range(1, len(training_loss) + 1)
plt.style.use('seaborn-whitegrid')
plt.rcParams['figure.dpi'] = 100
plt.rcParams['figure.figsize']=(7, 3.5)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.title('Loss function convergence', fontsize = 17)
plt.ylabel('Mean Squared Error', fontsize = 17)
plt.xlabel('Epoch', fontsize = 17)
plt.plot(epoch_count, training_loss)
```

```python
# Visualising results
days = range(1, len(predicted_exchange) + 1)
plt.style.use('seaborn-whitegrid')
plt.rcParams['figure.dpi'] = 100
plt.rcParams['figure.figsize']=(7, 3.5)
plt.title('Predicting USD/PLN exchange rate', fontsize = 17)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.ylabel('USD/PLN ratio', fontsize = 17)
plt.xlabel('Day', fontsize = 17)
plt.plot(days, test_values)
plt.plot(days, predicted_exchange)
plt.legend(['Real values', 'Predicted values'], prop={'size': 14})
plt.show()
```
